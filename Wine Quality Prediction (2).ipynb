Attributes related to the dataset

Fixed acidity - Fixed acidity is the set of the wine's natural acids.The predominant fixed acids found in wines are tartaric, malic, citric, and succinic.

Volatile acidity - Volatile acidity (VA) is a measure of the wine's volatile (or gaseous) acids. Acetic acid is the primary volatile acid in wine, but smaller amounts of lactic, formic, butyric, propionic acid, carbonic acid (from carbon dioxide), and sulfurous acid (from sulfur dioxide) may be present.

Citric acid - Amount of citric acid in wine.citric acid is found only in very minute quantities in wine grapes. It often has a concentration about 1/20 that of tartaric acid.

residual sugar - The amount of sugar remaining after fermentation stops.The residual sugar remaining in a dry wine is usually around 0.3 to 2 g/L.

chlorides - The amount of salt (Sodium chloride) in the wine.

free sulfur dioxide - The amount of free sulfur dioxide in the wine. Free sulfur dioxide protects wine by scavenging oxygen and interrupting microbiological activity is, itself, consumed in the process.

total sulfur dioxide - Amount of free and bound forms of SO2.

density - The measurement of how tightly a material is packed together.

pH - Describes how acidic or basic a wine is on a scale from 0 (very acidic) to 14 (very basic); most wines are between 3-4.

sulphites - Amount of sulphates present in wine.Sulphites in wine are used to stop fermentation at a specific point in the winemaking process.

alcohol - The percent alcohol content of the wine.

quality - Output variable.

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.preprocessing import StandardScaler
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

# Step 1: Load and Prepare the Data
  # Make sure to update this to the correct path if needed
wine_data = pd.read_csv('WineQT.csv')

# Create target classes
wine_data['quality_class'] = wine_data['quality'].apply(lambda x: 0 if x in [3, 4] else (1 if x in [5, 6] else 2))

# Features and new target variable
X = wine_data.drop(columns=['quality', 'Id', 'quality_class'])
y = wine_data['quality_class']

# Normalize (standardize) the features
scaler = StandardScaler()
X_normalized = scaler.fit_transform(X)

# Step 2: Split the data into training and testing sets with an 80/20 split
X_train, X_test, y_train, y_test = train_test_split(X_normalized, y, test_size=0.2, random_state=42, stratify=y)

# Step 3: Hyperparameter Tuning using GridSearchCV
# Define the parameter grid for SVM
param_grid_svm = {
    'C': [0.1, 1, 10, 100, 1000],
    'gamma': [1, 0.1, 0.01, 0.001, 0.0001],
    'kernel': ['rbf'],
    'class_weight': ['balanced', None]
}

# Set up GridSearchCV
svm_model = SVC(random_state=42)
grid_search_svm = GridSearchCV(
    estimator=svm_model,
    param_grid=param_grid_svm,
    scoring='accuracy',
    cv=5,
    verbose=2,
    n_jobs=-1
)

# Fit the model using GridSearchCV
grid_search_svm.fit(X_train, y_train)

# Get the best parameters and print them
best_params_svm = grid_search_svm.best_params_
print(f"Best Parameters for SVM: {best_params_svm}")

# Step 4: Train the Final SVM Model with the Best Parameters
# Apply the best parameters to the SVM model
svm_model_final = SVC(
    C=best_params_svm['C'],
    gamma=best_params_svm['gamma'],
    kernel=best_params_svm['kernel'],
    class_weight=best_params_svm['class_weight'],
    random_state=42
)

# Train the model
svm_model_final.fit(X_train, y_train)

# Predict on the test set
y_pred_svm = svm_model_final.predict(X_test)

# Calculate accuracy
accuracy_svm = accuracy_score(y_test, y_pred_svm)
print(f"SVM Test Accuracy: {accuracy_svm:.4f}")

# Generate the classification report
classification_report_svm = classification_report(y_test, y_pred_svm, target_names=['Bad Quality', 'Middle Quality', 'High Quality'])
print("SVM Classification Report:")
print(classification_report_svm)

# Confusion Matrix
conf_matrix_svm = confusion_matrix(y_test, y_pred_svm)
print("SVM Confusion Matrix:")
print(conf_matrix_svm)
```

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split, RandomizedSearchCV
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix



# Create target classes
wine_data['quality_class'] = wine_data['quality'].apply(lambda x: 0 if x in [3, 4] else (1 if x in [5, 6] else 2))

# Features and new target variable
X = wine_data.drop(columns=['quality', 'Id', 'quality_class'])
y = wine_data['quality_class']

# Normalize (standardize) the features
scaler = StandardScaler()
X_normalized = scaler.fit_transform(X)

# Step 2: Split the data into training and testing sets with an 80/20 split
X_train, X_test, y_train, y_test = train_test_split(X_normalized, y, test_size=0.2, random_state=42, stratify=y)

# Step 3: Hyperparameter Tuning using RandomizedSearchCV
# Define the parameter grid for Logistic Regression
param_grid_logistic = {
    'penalty': ['l1', 'l2', 'elasticnet', 'none'],
    'C': np.logspace(-4, 4, 20),
    'solver': ['lbfgs', 'newton-cg', 'liblinear', 'sag', 'saga'],
    'max_iter': [100, 1000, 2500, 5000]
}

# Set up RandomizedSearchCV
logistic_model = LogisticRegression()
random_search_logistic = RandomizedSearchCV(
    estimator=logistic_model,
    param_distributions=param_grid_logistic,
    cv=3,
    verbose=2,
    n_jobs=-1,
    random_state=42
)

# Fit RandomizedSearchCV to the data
random_search_logistic.fit(X_train, y_train)

# Get the best parameters and print them
best_params_logistic = random_search_logistic.best_params_
print(f"Best Parameters for Logistic Regression: {best_params_logistic}")

# Step 4: Train the Final Logistic Regression Model with the Best Parameters
# Apply the best parameters to Logistic Regression
logistic_model_final = LogisticRegression(
    C=best_params_logistic['C'],
    max_iter=best_params_logistic['max_iter'],
    penalty=best_params_logistic['penalty'],
    solver=best_params_logistic['solver'],
    random_state=42
)

# Train the final model
logistic_model_final.fit(X_train, y_train)

# Step 5: Evaluate the Final Model
# Predict on the test set
y_pred_logistic = logistic_model_final.predict(X_test)

# Calculate accuracy
accuracy_logistic = accuracy_score(y_test, y_pred_logistic)
print(f"Logistic Regression Test Accuracy: {accuracy_logistic:.4f}")

# Generate the classification report
classification_report_logistic = classification_report(y_test, y_pred_logistic, target_names=['Bad Quality', 'Middle Quality', 'High Quality'])
print("Logistic Regression Classification Report:")
print(classification_report_logistic)

# Confusion Matrix
conf_matrix_logistic = confusion_matrix(y_test, y_pred_logistic)
print("Logistic Regression Confusion Matrix:")
print(conf_matrix_logistic)
```

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split, RandomizedSearchCV
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix


# Create target classes
wine_data['quality_class'] = wine_data['quality'].apply(lambda x: 0 if x in [3, 4] else (1 if x in [5, 6] else 2))

# Features and new target variable
X = wine_data.drop(columns=['quality', 'Id', 'quality_class'])
y = wine_data['quality_class']

# Normalize (standardize) the features
scaler = StandardScaler()
X_normalized = scaler.fit_transform(X)

# Step 2: Split the data into training and testing sets with an 80/20 split
X_train, X_test, y_train, y_test = train_test_split(X_normalized, y, test_size=0.2, random_state=42, stratify=y)

# Step 3: Hyperparameter Tuning using RandomizedSearchCV
# Define the parameter grid for Random Forest
param_grid_rf = {
    'n_estimators': [100, 200, 500, 1000],  # Number of trees
    'max_depth': [None, 10, 20, 30, 40, 50],
    'min_samples_split': [2, 5, 10, 20],
    'min_samples_leaf': [1, 2, 4, 6],
    'bootstrap': [True, False]
}

# Set up RandomizedSearchCV
rf_model = RandomForestClassifier(random_state=42)
random_search_rf = RandomizedSearchCV(
    estimator=rf_model,
    param_distributions=param_grid_rf,
    n_iter=50,  # Number of random parameter settings to sample
    cv=3,
    verbose=2,
    n_jobs=-1,
    random_state=42
)

# Fit RandomizedSearchCV to the data
random_search_rf.fit(X_train, y_train)

# Get the best parameters and print them
best_params_rf = random_search_rf.best_params_
print(f"Best Parameters for Random Forest: {best_params_rf}")

# Step 4: Train the Final Random Forest Model with the Best Parameters
# Apply the best parameters to Random Forest
rf_model_final = RandomForestClassifier(
    n_estimators=best_params_rf['n_estimators'],
    max_depth=best_params_rf['max_depth'],
    min_samples_split=best_params_rf['min_samples_split'],
    min_samples_leaf=best_params_rf['min_samples_leaf'],
    bootstrap=best_params_rf['bootstrap'],
    random_state=42
)

# Train the final model
rf_model_final.fit(X_train, y_train)

# Step 5: Evaluate the Final Model
# Predict on the test set
y_pred_rf = rf_model_final.predict(X_test)

# Calculate accuracy
accuracy_rf = accuracy_score(y_test, y_pred_rf)
print(f"Random Forest Test Accuracy: {accuracy_rf:.4f}")

# Generate the classification report
classification_report_rf = classification_report(y_test, y_pred_rf, target_names=['Bad Quality', 'Middle Quality', 'High Quality'])
print("Random Forest Classification Report:")
print(classification_report_rf)

# Confusion Matrix
conf_matrix_rf = confusion_matrix(y_test, y_pred_rf)
print("Random Forest Confusion Matrix:")
print(conf_matrix_rf)
```

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split, RandomizedSearchCV
from sklearn.preprocessing import StandardScaler
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

# Step 1: Load and Prepare the Data
file_path = 'WineQT.csv'  # Make sure to update this to the correct path if needed
wine_data = pd.read_csv(file_path)

# Create target classes
wine_data['quality_class'] = wine_data['quality'].apply(lambda x: 0 if x in [3, 4] else (1 if x in [5, 6] else 2))

# Features and new target variable
X = wine_data.drop(columns=['quality', 'Id', 'quality_class'])
y = wine_data['quality_class']

# Normalize (standardize) the features
scaler = StandardScaler()
X_normalized = scaler.fit_transform(X)

# Step 2: Split the data into training and testing sets with an 80/20 split
X_train, X_test, y_train, y_test = train_test_split(X_normalized, y, test_size=0.2, random_state=42, stratify=y)

# Step 3: Hyperparameter Tuning using RandomizedSearchCV
# Define the parameter grid for Decision Tree
param_grid_dt = {
    'max_depth': [None, 10, 20, 30, 40, 50],
    'min_samples_split': [2, 5, 10, 20],
    'min_samples_leaf': [1, 2, 4, 6],
    'criterion': ['gini', 'entropy'],
    'splitter': ['best', 'random']
}

# Set up RandomizedSearchCV
dt_model = DecisionTreeClassifier(random_state=42)
random_search_dt = RandomizedSearchCV(
    estimator=dt_model,
    param_distributions=param_grid_dt,
    cv=3,
    n_iter=50,  # Number of parameter settings that are sampled
    verbose=2,
    n_jobs=-1,
    random_state=42
)

# Fit RandomizedSearchCV to the data
random_search_dt.fit(X_train, y_train)

# Get the best parameters and print them
best_params_dt = random_search_dt.best_params_
print(f"Best Parameters for Decision Tree: {best_params_dt}")

# Step 4: Train the Final Decision Tree Model with the Best Parameters
# Apply the best parameters to Decision Tree
dt_model_final = DecisionTreeClassifier(
    max_depth=best_params_dt['max_depth'],
    min_samples_split=best_params_dt['min_samples_split'],
    min_samples_leaf=best_params_dt['min_samples_leaf'],
    criterion=best_params_dt['criterion'],
    splitter=best_params_dt['splitter'],
    random_state=42
)

# Train the final model
dt_model_final.fit(X_train, y_train)

# Step 5: Evaluate the Final Model
# Predict on the test set
y_pred_dt = dt_model_final.predict(X_test)

# Calculate accuracy
accuracy_dt = accuracy_score(y_test, y_pred_dt)
print(f"Decision Tree Test Accuracy: {accuracy_dt:.4f}")

# Generate the classification report
classification_report_dt = classification_report(y_test, y_pred_dt, target_names=['Bad Quality', 'Middle Quality', 'High Quality'])
print("Decision Tree Classification Report:")
print(classification_report_dt)

# Confusion Matrix
conf_matrix_dt = confusion_matrix(y_test, y_pred_dt)
print("Decision Tree Confusion Matrix:")
print(conf_matrix_dt)
```
